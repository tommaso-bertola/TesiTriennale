```python
import numpy as np
from scipy import signal 
import matplotlib.pyplot as plt
```

```python
sample_rate=10 
nsamples=6000

a1=6
a2=12
b=0.9
c=0.35
d1=a1*b
d2=a2*b

tt=np.arange(300)/sample_rate
sig=np.zeros(300)
sig[0]=1

hrf=((tt/d1)**a1)*np.exp(-(tt-d1)/b)-c*((tt/d2)**a2)*np.exp(-(tt-d2)/b)
#plt.plot(tt, sig, color='g')
plt.plot(tt,hrf, color='r')
plt.xlabel("Tempo [s]")
plt.ylabel("Ampiezza [a.u.]")
#plt.title("Hemodynamic response function")
plt.axvline(0, linestyle='dashed', color='b')
plt.grid(True, linestyle='dotted')
plt.xlim(-1,25)
ax=plt.gca()
ax.text(0.4, 0.4, 'Evento scarica', rotation=90, va='center', color='b', fontsize=12)
ax.text(21, 0.03, 'Risposta\nemodinamica', ha='center', color='r', fontsize=12)

plt.savefig('../../Figure/hrf_tempo.pdf',bbox_inches='tight')
plt.show()
```
