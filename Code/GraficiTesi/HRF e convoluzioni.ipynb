```python
import numpy as np
from scipy import signal 
import matplotlib.pyplot as plt
```

```python
sample_rate=10 
nsamples=6000

a1=6
a2=12
b=0.9
c=0.35
d1=a1*b
d2=a2*b

tt=np.arange(300)/sample_rate
sig=np.zeros(300)
sig[0]=1

hrf=((tt/d1)**a1)*np.exp(-(tt-d1)/b)-c*((tt/d2)**a2)*np.exp(-(tt-d2)/b)
plt.figure(figsize=(7,4))
#plt.plot(tt, sig, color='g')
plt.plot(tt,hrf, color='r')
plt.xlabel("Tempo [s]")
plt.ylabel("Ampiezza [a.u.]")
#plt.title("Hemodynamic response function")
plt.axvline(0, linestyle='dashed', color='b')
plt.grid(True, linestyle='dotted')
plt.xlim(-1,25)
ax=plt.gca()
#ax.text(7,0.8,r'$h(t)=\left(\frac{t}{d_1}\right)^{a_1}\exp{\left(-\frac{t-d_1}{b_1}\right)}-c\left(\frac{t}{d_2}\right)^{a_2}\exp{\left(-\frac{t-d_2}{b_2}\right)}$',size=13)
#ax.text(10,0.6,r'$d_i=a_i b_i\,,\,a_1=6\,,\,a_2=12\,,\,b_i=0.9\,,\,c=0.35$',size=11, ha='left')
ax.text(0.4, 0.4, 'Evento scarica', rotation=90, va='center', color='b', fontsize=12)
ax.text(24.5, 0.04, 'Risposta\nemodinamica\nHRF', ha='right', color='r', fontsize=12)
#ax.text(10,0.8,r'HRF(t)=')
plt.savefig('../../FigurePowerPoint/hrf_tempo.svg',bbox_inches='tight', dpi=200)
plt.show()
```
