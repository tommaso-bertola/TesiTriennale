```python
# Plotta la matrice del connettoma con la colorbar

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.ticker as mticker
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
W66=np.loadtxt("../../Data/connectivity_matrix/weights.txt")
W998=np.loadtxt("../../Data/connectivity_matrix/group_mean_connectivity_matrix_file.txt")
```

```python
col = ["red", "blue"]
cmap1 = LinearSegmentedColormap.from_list("mycmap", col)
```

```python
plt.imshow(W66, cmap='cividis',norm=colors.LogNorm())
plt.colorbar(pad=+0.02)
#plt.clim(-1,1)
plt.axvline(x=32.5, color='black', alpha=0.7)
plt.axhline(y=32.5, color='black', alpha=0.7)
ax = plt.gca()

ax.yaxis.set_minor_locator(mticker.FixedLocator((66.0/4.0, 66/4.0*3)))
ax.yaxis.set_minor_formatter(mticker.FixedFormatter(("RH", "LH")))
ax.xaxis.set_minor_locator(mticker.FixedLocator((66.0/4.0, 66/4.0*3)))
ax.xaxis.set_minor_formatter(mticker.FixedFormatter(("RH", "LH")))

plt.setp(ax.yaxis.get_minorticklabels(), rotation=90, va="center")
ax.tick_params("y",which="minor",pad=25, left=False)
plt.setp(ax.xaxis.get_minorticklabels(), va="center")
ax.tick_params("x",which="minor",pad=25, left=False)


ax.set_yticklabels(ax.get_yticks(), rotation = 90)

ax.set_xticks([66/4,66/4*3])
ax.set_xticklabels(['LH','RH'])
ax.set_yticklabels([])


ax.set_yticks([66/4,66/4*3])
ax.set_yticklabels(['LH','RH'])
plt.savefig('../../FigurePowerPoint/connettoma_66_cividis.svg', bbox_inches='tight', dpi=200)


plt.show()
```

```python
plt.imshow(W998, cmap='cividis',norm=colors.LogNorm())
plt.colorbar()
#plt.clim(-1,1)
plt.axvline(x=998/2, color='black', alpha=0.7)
plt.axhline(y=998/2, color='black', alpha=0.7)
ax = plt.gca()

ax.yaxis.set_minor_locator(mticker.FixedLocator((998.0/4.0, 998/4.0*3)))
ax.yaxis.set_minor_formatter(mticker.FixedFormatter(("RH", "LH")))
ax.xaxis.set_minor_locator(mticker.FixedLocator((998.0/4.0, 998/4.0*3)))
ax.xaxis.set_minor_formatter(mticker.FixedFormatter(("RH", "LH")))

plt.setp(ax.yaxis.get_minorticklabels(), rotation=90, va="center")
ax.tick_params("y",which="minor",pad=25, left=False)
plt.setp(ax.xaxis.get_minorticklabels(), va="center")
ax.tick_params("x",which="minor",pad=25, left=False)
plt.savefig('../../Figure/connettoma_998_cividis.pdf', bbox_inches='tight')


plt.show()
```
