# Non normalizzato
fit delle dist cluster size **solo** per t=0.15 

```python
import matplotlib.ticker
import matplotlib.ticker
from scipy.optimize import curve_fit
import HTC
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pickle

output=pickle.load(open('not_norm_100run_6000time_0-0to0-3_compute_all_2.pickle','rb'))
```

```python
def fitter(x,alpha, gamma):
    return x**(-alpha)*np.exp(-x*gamma)


x,ll=HTC.get_sizes_distribution(output['s_distrib'])
```

```python
popt,_=curve_fit(fitter, x,ll)
alpha,gamma=3/2,1/66
x_model=np.linspace(x.min(),50, 100)
y_model=fitter(x_model, alpha, gamma)
#x,ll=HTC.get_sizes_distribution(output['s_distrib'])

fig, ax=plt.subplots()
ax.scatter(x,ll , color='b')
ax.plot(x_model, y_model, color='r')
ax.set_yscale("log")
ax.set_xscale("log")

ax.set_xlabel("Dimensione cluster")
ax.set_ylabel("p(dimensione)")
ax.set_xticks([1,2,4,8,16,24])
ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
plt.xlim(0.99,30)
plt.ylim(1e-5,2)
plt.grid(linestyle='dotted')
plt.title('Plot con alpha=3/2, gamma=1/66, QUELLO ATTUALE IN TESI')
plt.savefig('../Figure/Simulazione/Norm/dist_sizes.pdf', bbox_inches='tight')

plt.show()
```

```python
popt,_=curve_fit(fitter, x,ll)
alpha,gamma=popt
x_model=np.linspace(x.min(),50, 100)
y_model=fitter(x_model, alpha, gamma)
#x,ll=HTC.get_sizes_distribution(output['s_distrib'])

fig, ax=plt.subplots()
ax.scatter(x,ll , color='b')
ax.plot(x_model, y_model, color='r')
ax.set_yscale("log")
ax.set_xscale("log")

ax.set_xlabel("Dimensione cluster")
ax.set_ylabel("p(dimensione)")
ax.set_xticks([1,2,4,8,16,24])
ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
plt.xlim(0.99,30)
plt.ylim(1e-5,2)
plt.grid(linestyle='dotted')
plt.title('Plot con alpha='+str(popt[0])+', gamma='+str(popt[0])+', QUELLO FITTATO')
plt.savefig('../Figure/Simulazione/Norm/dist_sizes.pdf', bbox_inches='tight')

plt.show()
```
