```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
W=np.loadtxt("../Data/connectivity_matrix/weights.txt")
```

```python
w = W[W != 0].flatten()
w.shape
```

```python
w_max = W.max()+0.01
```

```python
hist, binning = np.histogram(w, bins=50, range=(w.min(), w_max), density=True)
```

```python
plt.bar(binning[:-1], hist ,width=np.diff(binning), log=True)
```

```python
cdf = np.cumsum(hist/hist.sum())
cdf
```

```python
plt.plot(cdf)
```

```python
u = np.random.uniform(0, 1, 20330)
u
```

```python
w_n = binning[np.searchsorted(cdf[1:-1], u)]
np.count_nonzero(w_n)
```

```python
np.sort(w_n)
```

```python
plt.hist(w_n, bins=50, log=True)
plt.show()
```
